```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#df = pd.DataFrame(data)
# Replace 'filename.xlsx' with your actual file path
df = pd.read_excel('Dataset.xlsx')
#print(df.head())
#print(df)
#'Curah Hujan (mm)',
columnNama = ['Penjualan A (pcs)','Penjualan B (pcs)']
for column in columnNama:
    print(f"Statistics for {column}:")
    
    # Calculate each statistic
    mean_value = df[column].mean()
    median_value = df[column].median()
    mode_value = df[column].mode()[0]  # Get the first mode
    range_value = df[column].max() - df[column].min()
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    variance_value = df[column].var()
    std_dev = df[column].std()
    skewness_value = df[column].skew()
    kurtosis_value = df[column].kurtosis()
    Q1 = df[column].quantile(0.25)
    Q2 = median_value  # Median is Q2
    Q3 = df[column].quantile(0.75)
    
    print(f"  Mean: {mean_value}")
    print(f"  Median: {median_value}")
    print(f"  Mode: {mode_value}")
    print(f"  Range: {range_value}")
    print(f"  Q1 (25th percentile): {Q1}")
    print(f"  Q2 (50th percentile, Median): {Q2}")
    print(f"  Q3 (75th percentile): {Q3}")
    print(f"  IQR: {IQR}")
    print(f"  Variance: {variance_value}")
    print(f"  Standard Deviation: {std_dev}")
    print(f"  Skewness: {skewness_value}")
    print(f"  Kurtosis: {kurtosis_value}")
    print("\n")
    print(f"Frequency Distribution for {column}:")
    
    # Frequency Distribution Table (using 'value_counts' or binning with 'cut')
    freq_table = pd.cut(df[column], bins=10).value_counts().sort_index()  # 10 bins as an example
    print(freq_table)
    print("\n")
    plt.figure(figsize=(8, 6))
    
    # Create histogram
    plt.hist(df[column], bins=10, alpha=0.7, color='blue', edgecolor='black')
    
    # Titles and labels
    plt.title(f'Histogram of {column}', fontsize=15)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.axvline(x=mean_value, color='green', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')
    plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')
    plt.axvline(x=Q1, color='orange', linestyle='dashed', linewidth=1, label=f'Q1 = {Q1:.2f}')
    plt.axvline(x=Q3, color='purple', linestyle='dashed', linewidth=1, label=f'Q3 = {Q3:.2f}')
    # Show plot
    plt.grid(True)
    plt.show()
    #for column in columns_to_analyze:
    plt.figure(figsize=(8, 6))
    
    # Create histogram data (without plotting)
    counts, bin_edges = np.histogram(df[column], bins=10)
    
    # Get the midpoints of the bins
    bin_midpoints = 0.5 * (bin_edges[1:] + bin_edges[:-1])
    
    # Plot frequency polygon (line plot)
    plt.plot(bin_midpoints, counts, marker='o', linestyle='-', color='blue')
    plt.axvline(x=mean_value, color='green', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')
    plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')
    plt.axvline(x=Q1, color='orange', linestyle='dashed', linewidth=1, label=f'Q1 = {Q1:.2f}')
    plt.axvline(x=Q3, color='purple', linestyle='dashed', linewidth=1, label=f'Q3 = {Q3:.2f}')
    
    # Titles and labels
    plt.title(f'Frequency Polygon of {column}', fontsize=15)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    
    # Show plot
    plt.grid(True)
    plt.show()
    plt.figure(figsize=(8, 6))
    
```
